```python
import requests
import json
import os

import pandas as pd
import matplotlib.pyplot as plt

from dotenv import load_dotenv

load_dotenv()
```

```python
key = os.environ.get('OPENWEATHERMAP_API_KEY')
city_name = 'Eindhoven'

request_url = f"https://api.openweathermap.org/data/2.5/weather?q={city_name}&appid={key}"
```

```python
import time

responses = []

for _ in range(5):
    res = requests.get(request_url)
    responses.append(res.json())
    
    # Sleep one second to avoid hitting the rate limit
    time.sleep(1)
```

```python
for response in responses:
    print(response)
    print()
```

```python
# weather is list of dict: id, main
# main is dict: temp, feels_like, temp_min, temp_max, pressure, humidity
# visibility
# wind is dict: speed, deg
# rain is dict: 1h
# clouds is dict: all
# dt is gonna be the index
# sys is dict: sunrine, sunset
columns = ['dt',
            'weather_0',
           'weather_1',  
           'temp', 
           'feels_like', 
           'temp_min', 
           'temp_max', 
           'pressure', 
           'humidity', 
           'visibility', 
           'wind_speed',
           'wind_deg' 
           'rain', 
           'clouds', 
           'sunrise', 
           'sunset']

dts = []
frames = []

for response in responses:
    weather_0 = response['weather'][0]['main']
    weather_1 = response['weather'][1]['main']
    
    temp = response['main']['temp']
    feels_like = response['main']['feels_like']
    temp_min = response['main']['temp_min']
    temp_max = response['main']['temp_max']
    pressure = response['main']['pressure']
    humidity = response['main']['humidity']
    
    visibility = response['visibility']
    
    wind_speed = response['wind']['speed']
    wind_deg = response['wind']['deg']
    
    rain = response['rain']['1h']
    
    clouds = response['clouds']['all']
    
    dt = response['dt']
    sunrise = response['sys']['sunrise']
    sunset = response['sys']['sunset']
    
    f = {
        'weather_0': weather_0,
        'weather_1': weather_1,
        'temp': temp,
        'feels_like': feels_like,
        'temp_min': temp_min,
        'temp_max': temp_max,
        'pressure': pressure,
        'humidity': humidity,
        'visibility': visibility,
        'wind_speed': wind_speed,
        'wind_deg': wind_deg,
        'rain': rain,
        'clouds': clouds,
        'sunrise': sunrise,
        'sunset': sunset
    }
    
    frames.append(f)
    dts.append(dt)
    
df = pd.DataFrame(frames, index=dts)
```
